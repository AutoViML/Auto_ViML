```python
import sys
print(sys.version)
import pandas as pd
import numpy as np
pd.options.display.max_columns=1000
datapath = '../data_sets/'
```

```python
target = ['casual','registered']
idcol = ''
sep=','
modeltype = 'Regression'
excl = ['cnt', 'temp'] 
```

```python
df = pd.read_csv(datapath+'bikeshare.csv',sep=sep,
                 header=0,
                 index_col=None)
df.drop(excl,axis=1,inplace=True)
print(df.shape)
df.head()
```

```python
num = int(0.9*df.shape[0])
train = df[:num]
test = df[num:]
sample_submission=''
scoring_parameter = ''
```

```python
from autoviml.Auto_ViML import Auto_ViML
#### If Boosting_Flag = True => XGBoost, False=>ExtraTrees, None=>Linear Model
model, features, trainm, testm = Auto_ViML(train, target, test,sample_submission='',hyper_param='GS',
            scoring_parameter='', Boosting_Flag=True, KMeans_Featurizer=False,
            Add_Poly=0, Stacking_Flag=False, Binning_Flag=False,
              Imbalanced_Flag=False, verbose=2)                                
```

```python
def accu(results, y_cv):
    return (results==y_cv).astype(int).sum(axis=0)/(y_cv.shape[0])
def rmse(results, y_cv):
    return np.sqrt(np.mean((results - y_cv)**2, axis=0))
```
